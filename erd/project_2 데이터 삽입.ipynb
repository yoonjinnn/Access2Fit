```python
%%sql
CREATE TABLE raw_data.floating_population (
    id INT IDENTITY(1,1) PRIMARY KEY,
    year_code INT,
    district_code INT,
    district_name VARCHAR(32),
    total BIGINT,
    male BIGINT,
    female BIGINT,
    age_10 BIGINT,
    age_20 BIGINT,
    age_30 BIGINT,
    age_40 BIGINT,
    age_50 BIGINT,
    age_60over BIGINT,
    time_0006 BIGINT,
    time_0611 BIGINT,
    time_1114 BIGINT,
    time_1417 BIGINT,
    time_1721 BIGINT,
    time_2124 BIGINT,
    mon BIGINT,
    tue BIGINT,
    wed BIGINT,
    thu BIGINT,
    fri BIGINT,
    sat BIGINT,
    sun BIGINT,
    FOREIGN KEY (district_code) REFERENCES raw_data.basic_info(district_code)
);
```
